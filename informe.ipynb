### Parte 3<a name="p3"></a>
<b>Letra:</b>
> Calcular la media, mediana, moda, rango, varianza y desviación estándar de las edades.

<b>Definiciones:</b>

- Media: Es el promedio aritmético de todas las edades en el conjunto de datos.
- Mediana: Es el valor medio de las edades cuando están ordenadas de menor a mayor.
- Moda: Es el valor más frecuente en las edades.
- Rango: Es la diferencia entre la mayor y la menor edad.
- Varianza: Mide cuánto varían las edades respecto a la media.
- Desviación Estándar: Es la raíz cuadrada de la varianza y proporciona una medida de dispersión de las edades.

<b>Desarrollo del código:</b>

Se utilizó el método describe() de pandas para calcular las estadísticas básicas como la media, mediana, varianza y desviación estándar. Luego, se calculó la moda y el rango de manera independiente utilizando funciones específicas de pandas y numpy.

A continuación, se muestra el código utilizado para calcular y mostrar estas estadísticas:


```python
# Calcular estadísticas descriptivas básicas
age_descriptive_stats = titanic_df['age'].describe()

# Calcular la moda
mode_age = titanic_df['age'].mode().iloc[0]

# Calcular el rango
range_age = titanic_df['age'].max() - titanic_df['age'].min()

# Calcular la varianza
variance_age = titanic_df['age'].var()

# Crear un nuevo DataFrame con las estadísticas ordenadas
ordered_stats_df = pd.DataFrame({
    'mean': age_descriptive_stats['mean'],
    '50%': age_descriptive_stats['50%'],
    'mode': mode_age,
    'range': range_age,
    'variance': variance_age,
    'std': age_descriptive_stats['std'],
}, index=[0])

# Mostrar las estadísticas en el orden
print("Estadísticas descriptivas de las edades:")
for index, row in ordered_stats_df.iterrows():
    for stat, value in row.items():
        print(f"{stat}: {value}")
```

<b>Resultados obtenidos:</b>

- Media (mean): 57.5279683770243
- Mediana (50%): 57.58151073529475
- Moda (mode): 1.0977265142380197
- Rango (range): 98.86854607597618
- Varianza (variance): 630.0389016040109
- Desviación estandar (std): 25.100575722560844

Se observa que la media y la mediana son valores cercanos, lo cual sugiere que la distribución de las edades podría estar centrada alrededor de esos valores. Además, el rango muestra la amplitud total de las edades observadas en el conjunto de datos.

### Parte 4<a name="p4"></a>
<b>Letra:</b>
> ¿Cuál es la tasa de supervivencia general?

<b>Procedimiento:</b>

Para calcular la tasa de supervivencia general, se siguieron los siguientes pasos:
- Paso 1: se obtuvo el número total de pasajeros en el conjunto de datos contando el número total de filas.
- Paso 2: se contaron cuántos pasajeros sobrevivieron sumando los valores en la columna suvived donde 1 indica que el pasajero sobrevivió.
- Paso 3: se calculó la tasa de supervivencia dividiendo el número de sobrevivientes entre el número total de pasajeros y multiplicando por 100 para obtener el porcentaje.

A continuación, se muestra el cálculo en Python:

```python
# Calcular la tasa de supervivencia
total_passengers = len(titanic_df)
total_survivors = titanic_df['survived'].sum()
survival_rate = (total_survivors / total_passengers) * 100

# Mostrar la tasa de supervivencia
print(f"Tasa de supervivencia general: {survival_rate:.2f}%")
```

<b>Resultado obtenido:</b>

La tasa de supervivencia general del conjunto de datos del Titanic es del 40.09%, esto significa que el 40.09% de todos los pasajeros incluidos en el análisis que sobrevivieron al desastre.

### Parte 5<a name="p5"></a>
<b>Letra:</b>
> ¿Cuál es la tasa de supervivencia por género?

<b>Procedimiento:</b>

Para calcular la tasa de supervivencia general, se siguieron los siguientes pasos:
- Paso 1: se obtuvo el número total de pasajeros por género en el conjunto de datos.
- Paso 2: se contaron cuántos pasajeros sobrevivieron por género.
- Paso 3: se calculó la tasa de supervivencia por género dividiendo el número de sobrevivientes de cada género entre el número total de pasajeros de ese género y multiplicando por 100 para obtener el porcentaje.

A continuación, se muestra el cálculo en Python:

```python
# Calcular el total de pasajeros y sobrevivientes por género
total_passengers_gender = titanic_df['gender'].value_counts()
total_survivors_gender = titanic_df.groupby('gender')['survived'].sum()

# Calcular la tasa de supervivencia por género
survival_rate_gender = (total_survivors_gender / total_passengers_gender) * 100

# Mostrar la tasa de supervivencia por género
print(f"Tasa de supervivencia por género:")
print(survival_rate_gender)
```

<b>Resultado obtenido:</b>

La tasa de supervivencia por género es la siguiente:
- Femenino: 47.789474%
- Masculino: 32.579651%

Esto indica que aproximadamente el 47.789474% de las mujeres sobrevivieron, mientras que el 32.579651% de los hombres lograron sobrevivir. Este resultado refleja la prioridad dada a las mujeres y niños durante el rescate del Titanic.

<b>Modelo propuesto para la distribución de la variable edad:</b>

Como propuesta de un modelo para la distribución de edad, se ajustó una distribución normal a las edades de todos los pasajeros usando la biblioteca scipy:

```python
from cv2 import norm
from matplotlib import pyplot as plt
import numpy as np

# Ajustar una distribución normal a las edades usando scipy
ages_all = titanic_df['age'].dropna()
mu, std = norm.fit(ages_all)

# Crear una secuencia de valores para la distribución ajustada
xmin, xmax = min(ages_all), max(ages_all)
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)

# Dibujar el histograma y la distribución ajustada
plt.figure(figsize=(8, 6))
plt.hist(ages_all, bins=20, density=True, alpha=0.6, color='g', edgecolor='black')
plt.plot(x, p, 'k', linewidth=2)
plt.title('Distribución de la Edad con Ajuste Normal (Scipy)')
plt.xlabel('Edad')
plt.ylabel('Densidad de Frecuencia')
plt.show()

print(f"Parámetros de la distribución normal ajustada: media = {mu:.2f}, desviación estándar = {std:.2f}")
```

En este gráfico:
- El histograma verde representa la distribución de las edades de los pasajeros.
- La línea negra muestra la distribución ajustada a una distribución normal, utilizando los parámetros de media y desviación estándar obtenidos del ajuste.

Este modelo propuesto sugiere que las edades de los pasajeros siguen una distribución normal
